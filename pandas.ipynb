```python
def em(x):
  if x == 'S':
    return 1
  elif x == 'C':
    return 2
  elif x == 'Q':
    return 3
  else:
    return 0   
```

```python
df['em-c'] = df['Embarked'].apply(em)

df.head()
```

```python
df['em-c'] = df['Embarked'].apply(lambda x:  1 if x=='S' else (2 if x=='C' else (3 if x=='Q' else 0)))

df.head()
```
